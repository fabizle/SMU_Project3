```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data and check first five roads
df = pd.read_csv("Resources/tornados.csv")
df.head()
```

```python
df.shape
```

```python
# Check data type and missing values
df.info()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# Drop duplicate row
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
# Check missing values by column
df.isna().sum()
```

```python
# Check the missing values for'mag'
df["mag"].value_counts(dropna = False)
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
df['mag'].isnull().sum()
```

```python
df['loss'].isnull().sum()
```

```python
# Another check missing values by column
df.isna().sum()
```

```python
df["st"].unique()
```

```python
df.st.value_counts().plot(kind='bar', color=sns.color_palette('crest'), figsize=(10, 7))
plt.ylabel("Count")
plt.xlabel("State")
plt.title("Number of Tornadoes by State")
```

```python
# Plot tornados by year and magnitude
data= df.groupby(["st", "mag"])["st"].count().reset_index(name = 'Count')
plt.figure(figsize = (18, 8)) 
sns.lineplot(data=data, x="st", y="Count", hue='mag', marker="o", palette = 'crest')
plt.title("Tornado Magnitudes by State")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```
